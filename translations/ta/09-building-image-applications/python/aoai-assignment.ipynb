## DALL-E மற்றும் Midjourney எப்படி செயல்படுகின்றன

முதலில், [DALL-E](https://arxiv.org/pdf/2102.12092.pdf?WT.mc_id=academic-105485-koreyst). DALL-E என்பது transformer architecture அடிப்படையிலான Generative AI மாடல், இது *autoregressive transformer* கொண்டது.

ஒரு *autoregressive transformer* என்பது ஒரு மாடல் உரை விளக்கங்களிலிருந்து படங்களை உருவாக்கும் முறையை வரையறுக்கிறது. இது ஒரு பிக்சலை ஒரே நேரத்தில் உருவாக்குகிறது, பின்னர் உருவாக்கப்பட்ட பிக்சல்களை அடுத்த பிக்சலை உருவாக்க பயன்படுத்துகிறது. இது ஒரு நரம்பு வலையமைப்பில் பல அடுக்குகள் வழியாக செல்கிறது, படம் முழுமையாக உருவாகும் வரை.

இந்த செயல்முறையின் மூலம், DALL-E உருவாக்கும் படத்தில் பண்புகள், பொருட்கள், குணங்கள் மற்றும் பலவற்றை கட்டுப்படுத்துகிறது. ஆனால், DALL-E 2 மற்றும் 3 உருவாக்கப்படும் படத்தின் மீது மேலும் கட்டுப்பாட்டை வழங்குகின்றன.


## உங்கள் முதல் பட உருவாக்க பயன்பாட்டை உருவாக்குதல்

பட உருவாக்க பயன்பாட்டை உருவாக்க என்ன தேவை? உங்களுக்கு பின்வரும் நூலகங்கள் தேவை:

- **python-dotenv**, உங்கள் ரகசியங்களை *.env* கோப்பில் இருந்து விலக்கி வைக்க இந்த நூலகத்தை பயன்படுத்த பரிந்துரைக்கப்படுகிறது.
- **openai**, OpenAI API உடன் தொடர்பு கொள்ள இந்த நூலகத்தை நீங்கள் பயன்படுத்துவீர்கள்.
- **pillow**, Python-இல் படங்களை செயல்படுத்த.
- **requests**, HTTP கோரிக்கைகளை செய்ய உதவ.

1. பின்வரும் உள்ளடக்கத்துடன் *.env* என்ற கோப்பை உருவாக்கவும்:

    ```text
    AZURE_OPENAI_ENDPOINT=<your endpoint>
    AZURE_OPENAI_API_KEY=<your key>
    ```

    உங்கள் Azure Portal-இல் "Keys and Endpoint" பிரிவில் உங்கள் வளத்திற்கான இந்த தகவலை கண்டறியவும்.


1. மேலே உள்ள நூலகங்களை *requirements.txt* என்ற கோப்பில் சேகரிக்கவும், இதைப் போல:

    ```text
    python-dotenv
    openai
    pillow
    requests
    ```


1. அதன் பிறகு, ஒரு மெய்நிகர் சூழலை உருவாக்கி நூலகங்களை நிறுவவும்:


````python
# create virtual env
! python3 -m venv venv
# activate environment
! source venv/bin/activate
# install libraries
# pip install -r requirements.txt, if using a requirements.txt file 
! pip install python-dotenv openai pillow requests
````


1. *app.py* என்ற கோப்பில் பின்வரும் குறியீட்டை சேர்க்கவும்:

    ```python
    import openai
    import os
    import requests
    from PIL import Image
    import dotenv
    
    # import dotenv
    dotenv.load_dotenv()
    
    # Get endpoint and key from environment variables
    openai.api_base = os.environ['AZURE_OPENAI_ENDPOINT']
    openai.api_key = os.environ['AZURE_OPENAI_API_KEY']     
    
    # Assign the API version (DALL-E is currently supported for the 2023-06-01-preview API version only)
    openai.api_version = '2023-06-01-preview'
    openai.api_type = 'azure'
    
    
    try:
        # Create an image by using the image generation API
        generation_response = openai.Image.create(
            prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # Enter your prompt text here
            size='1024x1024',
            n=2,
            temperature=0,
        )
        # Set the directory for the stored image
        image_dir = os.path.join(os.curdir, 'images')
    
        # If the directory doesn't exist, create it
        if not os.path.isdir(image_dir):
            os.mkdir(image_dir)
    
        # Initialize the image path (note the filetype should be png)
        image_path = os.path.join(image_dir, 'generated-image.png')
    
        # Retrieve the generated image
        image_url = generation_response["data"][0]["url"]  # extract image URL from response
        generated_image = requests.get(image_url).content  # download the image
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)
    
        # Display the image in the default image viewer
        image = Image.open(image_path)
        image.show()
    
    # catch exceptions
    except openai.InvalidRequestError as err:
        print(err)

    ```

இந்த குறியீட்டை விளக்குவோம்:

- முதலில், OpenAI நூலகம், dotenv நூலகம், requests நூலகம் மற்றும் Pillow நூலகம் போன்ற தேவையான நூலகங்களை இறக்குமதி செய்கிறோம்.

    ```python

    ```

- அடுத்ததாக, *.env* கோப்பிலிருந்து சூழல் மாறிகளை ஏற்றுகிறோம்.

    ```python

    ```

- அதன் பிறகு, OpenAI API-க்கான endpoint, key, version மற்றும் type ஆகியவற்றை அமைக்கிறோம்.

    ```python

    ```

- பின்னர், படத்தை உருவாக்குகிறோம்:

    ```python

    ```

    மேலே உள்ள குறியீடு உருவாக்கப்பட்ட படத்தின் URL-ஐ கொண்ட JSON பொருளை பதிலளிக்கிறது. இந்த URL-ஐ பயன்படுத்தி படத்தை பதிவிறக்கம் செய்து கோப்பாக சேமிக்கலாம்.

- கடைசியாக, படத்தை திறந்து, தரப்பட்ட image viewer-ஐ பயன்படுத்தி அதை காண்பிக்கிறோம்:

```python
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # Enter your prompt text here
        size='1024x1024',
        n=2,
        temperature=0,
    )
```

- **prompt**, என்பது படத்தை உருவாக்க பயன்படுத்தப்படும் உரை prompt ஆகும். இங்கு, "Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils" என்ற prompt பயன்படுத்தப்படுகிறது.
- **size**, உருவாக்கப்படும் படத்தின் அளவை குறிக்கிறது. இங்கு, 1024x1024 பிக்சல்கள் அளவிலான படம் உருவாக்கப்படுகிறது.
- **n**, உருவாக்கப்படும் படங்களின் எண்ணிக்கையை குறிக்கிறது. இங்கு, இரண்டு படங்கள் உருவாக்கப்படுகின்றன.
- **temperature**, Generative AI மாதிரியின் வெளியீட்டின் சீர்மையைக் கட்டுப்படுத்தும் அளவுரு. இது 0 மற்றும் 1 இடையிலான மதிப்பாக இருக்கும், 0 என்றால் வெளியீடு நிர்ணயமாக இருக்கும், 1 என்றால் வெளியீடு சீரற்றமாக இருக்கும். இயல்புநிலை மதிப்பு 0.7 ஆகும்.

படங்களுடன் மேலும் செய்யக்கூடிய விஷயங்களை அடுத்த பகுதியில் காண்போம்.

## பட உருவாக்கத்தின் கூடுதல் திறன்கள்

Python-ல் சில வரிகளின் மூலம் படத்தை உருவாக்கியதை இதுவரை பார்த்தீர்கள். ஆனால், படங்களுடன் மேலும் பல விஷயங்களை செய்யலாம்.

நீங்கள் இதையும் செய்யலாம்:

- **திருத்தங்களைச் செய்யவும்**. ஏற்கனவே உள்ள படத்திற்கும், ஒரு மாஸ்க்கிற்கும், ஒரு prompt-க்கும் மூலம் படத்தை மாற்றலாம். உதாரணமாக, ஒரு படத்தின் ஒரு பகுதியை மாற்றலாம். நம் குரங்கு படத்தை கற்பனை செய்யுங்கள், குரங்கிற்கு ஒரு தொப்பி சேர்க்கலாம். இதைச் செய்ய, படத்தை, மாஸ்க்கை (மாற்றம் செய்ய வேண்டிய பகுதியை அடையாளம் காட்டும்) மற்றும் உரை prompt-ஐ வழங்க வேண்டும்.

    ```python
    response = openai.Image.create_edit(
      image=open("base_image.png", "rb"),
      mask=open("mask.png", "rb"),
      prompt="An image of a rabbit with a hat on its head.",
      n=1,
      size="1024x1024"
    )
    image_url = response['data'][0]['url']
    ```

    அடிப்படை படம் குரங்கையே மட்டும் கொண்டிருக்கும், ஆனால் இறுதி படம் குரங்கிற்கு தொப்பியுடன் இருக்கும்.
