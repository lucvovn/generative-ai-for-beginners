# உரை உருவாக்க பயன்பாடுகளை உருவாக்குதல்

இந்த பாடத்திட்டத்தின் மூலம் நீங்கள் இதுவரை முக்கியமான கருத்துக்களைப் பார்த்துள்ளீர்கள், உதாரணமாக "ப்ராம்ப்ட்" மற்றும் "ப்ராம்ப்ட் இன்ஜினியரிங்" எனப்படும் ஒரு முழு துறை. ChatGPT, Office 365, Microsoft Power Platform போன்ற பல கருவிகளை நீங்கள் பயன்படுத்தி ப்ராம்ப்ட்களை கொண்டு ஒரு செயல்பாட்டை நிறைவேற்ற முடியும்.

இந்த அனுபவத்தை ஒரு பயன்பாட்டில் சேர்க்க நீங்கள் ப்ராம்ப்ட்கள், கம்ப்ளீஷன்கள் போன்ற கருத்துகளைப் புரிந்து கொள்ள வேண்டும் மற்றும் ஒரு நூலகத்தைத் தேர்ந்தெடுக்க வேண்டும். இதுதான் இந்த அத்தியாயத்தில் நீங்கள் கற்றுக்கொள்ளப் போகிறீர்கள்.

## அறிமுகம்

இந்த அத்தியாயத்தில், நீங்கள்:

- openai நூலகம் மற்றும் அதன் முக்கியமான கருத்துகளைப் பற்றி கற்றுக்கொள்வீர்கள்.
- openai பயன்படுத்தி ஒரு உரை உருவாக்க பயன்பாட்டை உருவாக்குவீர்கள்.
- ப்ராம்ப்ட், டெம்பரேச்சர், மற்றும் டோக்கன்கள் போன்ற கருத்துகளைப் பயன்படுத்தி ஒரு உரை உருவாக்க பயன்பாட்டை உருவாக்குவது எப்படி என்பதைப் புரிந்து கொள்வீர்கள்.

## கற்றல் இலக்குகள்

இந்த பாடத்தின் முடிவில், நீங்கள்:

- ஒரு உரை உருவாக்க பயன்பாடு என்ன என்பதை விளக்க முடியும்.
- openai பயன்படுத்தி ஒரு உரை உருவாக்க பயன்பாட்டை உருவாக்க முடியும்.
- உங்கள் பயன்பாட்டை அதிகமாக அல்லது குறைவாக டோக்கன்களை பயன்படுத்தவும், மேலும் டெம்பரேச்சரை மாற்றவும் அமைக்க முடியும், பல்வேறு வெளியீடுகளுக்காக.

## உரை உருவாக்க பயன்பாடு என்ன?

சாதாரணமாக நீங்கள் ஒரு பயன்பாட்டை உருவாக்கும்போது, அது பின்வருமாறு ஒரு வகையான இடைமுகத்தைக் கொண்டிருக்கும்:

- **கமாண்ட் அடிப்படையிலானது**. Console பயன்பாடுகள் பொதுவாக நீங்கள் ஒரு கமாண்டை டைப் செய்து ஒரு பணியை நிறைவேற்றும் பயன்பாடுகள். உதாரணமாக, `git` என்பது ஒரு கமாண்ட் அடிப்படையிலான பயன்பாடு.
- **பயனர் இடைமுகம் (UI)**. சில பயன்பாடுகள் கிராபிகல் பயனர் இடைமுகங்களை (GUIs) கொண்டிருக்கும், இதில் நீங்கள் பட்டன்களை கிளிக் செய்வது, உரையை உள்ளிடுவது, விருப்பங்களைத் தேர்ந்தெடுப்பது போன்றவை செய்யலாம்.

### Console மற்றும் UI பயன்பாடுகள் வரையறுக்கப்பட்டவை

ஒரு கமாண்ட் அடிப்படையிலான பயன்பாட்டுடன் ஒப்பிடுக:

- **இது வரையறுக்கப்பட்டது**. நீங்கள் எந்த கமாண்டையும் டைப் செய்ய முடியாது, பயன்பாடு ஆதரிக்கும் கமாண்ட்களையே மட்டும் டைப் செய்ய முடியும்.
- **மொழி குறிப்பிட்டது**. சில பயன்பாடுகள் பல மொழிகளை ஆதரிக்கின்றன, ஆனால் இயல்பாக பயன்பாடு ஒரு குறிப்பிட்ட மொழிக்காக உருவாக்கப்பட்டுள்ளது, கூடுதல் மொழி ஆதரவைச் சேர்க்க முடிந்தாலும்.

### உரை உருவாக்க பயன்பாடுகளின் நன்மைகள்

அப்படியென்றால், ஒரு உரை உருவாக்க பயன்பாடு எப்படி வேறுபடுகிறது?

ஒரு உரை உருவாக்க பயன்பாட்டில், நீங்கள் அதிக சுதந்திரம் பெறுகிறீர்கள், ஒரு கமாண்ட்களின் தொகுப்புக்கு அல்லது ஒரு குறிப்பிட்ட உள்ளீடு மொழிக்கு மட்டுமே வரையறுக்கப்படவில்லை. அதற்கு பதிலாக, நீங்கள் இயற்கை மொழியைப் பயன்படுத்தி பயன்பாட்டுடன் தொடர்பு கொள்ளலாம். மேலும், நீங்கள் ஏற்கனவே ஒரு பெரிய தகவல் தொகுப்பில் பயிற்சி பெற்ற தரவுத் மூலத்துடன் தொடர்பு கொள்கிறீர்கள், ஆனால் ஒரு பாரம்பரிய பயன்பாடு ஒரு தரவுத்தொகுப்பில் உள்ளவற்றில் மட்டுமே வரையறுக்கப்பட்டிருக்கலாம்.

### உரை உருவாக்க பயன்பாட்டை கொண்டு என்ன உருவாக்க முடியும்?

நீங்கள் பலவற்றை உருவாக்க முடியும். உதாரணமாக:

- **ஒரு சாட்பாட்**. உங்கள் நிறுவனம் மற்றும் அதன் தயாரிப்புகள் போன்ற தலைப்புகள் குறித்து கேள்விகளுக்கு பதிலளிக்கும் ஒரு சாட்பாட் ஒரு நல்ல பொருத்தமாக இருக்கலாம்.
- **உதவியாளர்**. LLMக்கள் உரையை சுருக்குதல், உரையிலிருந்து தகவல்களைப் பெறுதல், ரெஸ்யூமே போன்ற உரைகளை உருவாக்குதல் போன்றவற்றில் சிறந்தவை.
- **கோடு உதவியாளர்**. நீங்கள் பயன்படுத்தும் மொழி மாதிரியைப் பொறுத்து, கோடு எழுத உதவும் ஒரு கோடு உதவியாளரை உருவாக்க முடியும். உதாரணமாக, GitHub Copilot மற்றும் ChatGPT போன்ற தயாரிப்புகளை நீங்கள் கோடு எழுத உதவியாக பயன்படுத்தலாம்.

## நான் எப்படி தொடங்கலாம்?

நீங்கள் ஒரு LLM உடன் ஒருங்கிணைக்க ஒரு வழியை கண்டுபிடிக்க வேண்டும், இது பொதுவாக இரண்டு அணுகுமுறைகளை உள்ளடக்கியது:

- **API பயன்படுத்துதல்**. இங்கு நீங்கள் உங்கள் ப்ராம்ப்டுடன் வலை கோரிக்கைகளை உருவாக்கி, உருவாக்கப்பட்ட உரையை திரும்பப் பெறுகிறீர்கள்.
- **நூலகம் பயன்படுத்துதல்**. நூலகங்கள் API அழைப்புகளைச் சுருக்கி, அவற்றைப் பயன்படுத்த எளிதாக்குகின்றன.

LLMக்களுடன் வேலை செய்ய சில பிரபலமான நூலகங்கள் உள்ளன:

- **openai**, இந்த நூலகம் உங்கள் மாதிரியை இணைத்து ப்ராம்ப்ட்களை அனுப்ப எளிதாக்குகிறது.

பிறகு, மேல்நிலையான செயல்பாடுகளை வழங்கும் நூலகங்கள் உள்ளன:

## openai பயன்படுத்தி முதல் பயன்பாடு

நாம் எவ்வாறு முதல் பயன்பாட்டை உருவாக்கலாம், எந்த நூலகங்கள் தேவை, எவ்வளவு தேவை என்பதைக் காண்போம்.

### openai நிறுவுதல்

  > [!NOTE] இந்த படி Codespaces அல்லது Devcontainer-ல் இந்த நோட்புக் இயக்கும்போது தேவையில்லை

OpenAI அல்லது Azure OpenAI உடன் தொடர்பு கொள்ள பல நூலகங்கள் உள்ளன. C#, Python, JavaScript, Java மற்றும் பல போன்ற பல நிரலாக்க மொழிகளைப் பயன்படுத்த முடியும்.  
நாம் `openai` Python நூலகத்தைத் தேர்ந்தெடுத்துள்ளோம், எனவே `pip` பயன்படுத்தி அதை நிறுவுவோம்.

```bash
pip install openai
```

இந்த நோட்புக் Codespaces அல்லது Dev Container-ல் இயக்கப்படவில்லை என்றால், உங்கள் கணினியில் [Python](https://www.python.org/) ஐயும் நிறுவ வேண்டும்.

### ஒரு வளத்தை உருவாக்கி உங்கள் API கீயை கண்டறியவும்

நீங்கள் இதுவரை செய்யவில்லை என்றால், பின்வரும் படிகளைச் செய்ய வேண்டும்:

- OpenAI-ல் ஒரு கணக்கை உருவாக்கவும் <https://platform.openai.com/signup>.
- இப்போது, உங்கள் API கீய்களைப் பெறவும் <https://platform.openai.com/api-keys>. 

>[!NOTE]
> உங்கள் API கீயை உங்கள் கோடில் இருந்து பிரிக்க வேண்டும். நீங்கள் அதை சூழல் மாறிகள் மூலம் செய்யலாம்.
> - உங்கள் .env கோப்பில் `OPENAI_KEY` சூழல் மாறியை உங்கள் API கீயாக அமைக்கவும். இந்த பாடத்திட்டத்தின் முந்தைய பயிற்சிகளை நீங்கள் முடித்திருந்தால், நீங்கள் அனைத்தும் அமைத்துவிட்டீர்கள்.
> - API கீயை ஒருமுறை மட்டுமே அணுக முடியும் என்பதை கவனிக்க வேண்டும். எனவே, அது சரியாக நகலெடுக்கப்பட்டதா என்பதை உறுதிப்படுத்துவது அவசியம். அது எதிர்பார்த்தபடி செயல்படவில்லை என்றால், கீயை நீக்கி புதிய ஒன்றை உருவாக்க பரிந்துரைக்கப்படுகிறது.

### OpenAI அமைப்பை அமைத்தல்

நீங்கள் OpenAI பயன்படுத்தினால், அமைப்பை அமைப்பது எப்படி என்பதை இங்கே காணலாம்:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-3.5-turbo"
```

மேலே நாம் பின்வருவனவற்றை அமைக்கிறோம்:

- `api_key`, இது OpenAI டாஷ்போர்டில் காணப்படும் உங்கள் API கீ.
- `deployment`, இது உங்கள் GPT பதிப்பு.

> [!NOTE]
> `os.environ` என்பது சூழல் மாறிகளைப் படிக்கும் ஒரு செயல்பாடாகும். `OPENAI_API_KEY` போன்ற சூழல் மாறிகளைப் படிக்க அதை நீங்கள் பயன்படுத்தலாம்.

## உரை உருவாக்குதல்

உரை உருவாக்குவதற்கான வழி `chat.completion` வகுப்பைப் பயன்படுத்துவது. இதோ ஒரு உதாரணம்:

```python
prompt = "Complete the following: Once upon a time there was a"

completion = client.chat.completions.create(model=deployment, messages=[{"role": "user", "content": prompt}])
print(completion.choices[0].message.content)
```
