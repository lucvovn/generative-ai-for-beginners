# உரை உருவாக்க பயன்பாடுகளை உருவாக்குதல்

இந்த பாடத்திட்டத்தின் மூலம் நீங்கள் இதுவரை முக்கியமான கருத்துக்களைப் பார்த்துள்ளீர்கள், உதாரணமாக "ப்ராம்ப்ட்" மற்றும் "ப்ராம்ப்ட் என்ஜினியரிங்" எனப்படும் ஒரு முழு துறை. ChatGPT, Office 365, Microsoft Power Platform போன்ற பல கருவிகளைப் பயன்படுத்தி, ப்ராம்ப்ட்களை பயன்படுத்தி ஒரு செயல்பாட்டை முடிக்க முடியும்.

இந்த அனுபவத்தை ஒரு பயன்பாட்டில் சேர்க்க நீங்கள் ப்ராம்ப்ட்கள், கம்ப்ளீஷன்கள் போன்ற கருத்துகளைப் புரிந்துகொள்ள வேண்டும் மற்றும் வேலை செய்ய ஒரு நூலகத்தைத் தேர்ந்தெடுக்க வேண்டும். இந்த அத்தியாயத்தில் நீங்கள் இதையே கற்றுக்கொள்வீர்கள்.

## அறிமுகம்

இந்த அத்தியாயத்தில், நீங்கள்:

- openai நூலகம் மற்றும் அதன் முக்கியமான கருத்துகளைப் பற்றி கற்றுக்கொள்வீர்கள்.
- openai பயன்படுத்தி ஒரு உரை உருவாக்க பயன்பாட்டை உருவாக்குவீர்கள்.
- ப்ராம்ப்ட், வெப்பநிலை, மற்றும் டோக்கன்கள் போன்ற கருத்துகளைப் பயன்படுத்தி ஒரு உரை உருவாக்க பயன்பாட்டை உருவாக்குவது எப்படி என்பதைப் புரிந்துகொள்வீர்கள்.

## கற்றல் இலக்குகள்

இந்த பாடத்தின் முடிவில், நீங்கள்:

- ஒரு உரை உருவாக்க பயன்பாடு என்ன என்பதை விளக்க முடியும்.
- openai பயன்படுத்தி ஒரு உரை உருவாக்க பயன்பாட்டை உருவாக்க முடியும்.
- உங்கள் பயன்பாட்டை அதிகமாக அல்லது குறைவாக டோக்கன்களைப் பயன்படுத்தவும், மேலும் வெப்பநிலையை மாற்றவும், பல்வேறு வெளியீடுகளை பெறவும் அமைக்க முடியும்.

## உரை உருவாக்க பயன்பாடு என்ன?

சாதாரணமாக, நீங்கள் ஒரு பயன்பாட்டை உருவாக்கும்போது, அது பின்வருமாறு ஒரு வகையான இடைமுகத்தைக் கொண்டிருக்கும்:

- **கமாண்ட் அடிப்படையிலானது**. Console பயன்பாடுகள் பொதுவாக நீங்கள் ஒரு கமாண்டை டைப் செய்து ஒரு பணியைச் செய்யும் பயன்பாடுகள். உதாரணமாக, `git` என்பது ஒரு கமாண்ட் அடிப்படையிலான பயன்பாடு.
- **பயனர் இடைமுகம் (UI)**. சில பயன்பாடுகள் கிராபிகல் பயனர் இடைமுகங்களை (GUIs) கொண்டிருக்கும், இதில் நீங்கள் பட்டன்களை கிளிக் செய்யலாம், உரை உள்ளிடலாம், விருப்பங்களைத் தேர்ந்தெடுக்கலாம் மற்றும் பல.

### Console மற்றும் UI பயன்பாடுகள் வரையறுக்கப்பட்டவை

ஒரு கமாண்ட் அடிப்படையிலான பயன்பாட்டுடன் ஒப்பிடுக:

- **இது வரையறுக்கப்பட்டது**. நீங்கள் எந்த கமாண்டையும் டைப் செய்ய முடியாது, பயன்பாடு ஆதரிக்கும் கமாண்ட்களையே மட்டும் டைப் செய்ய முடியும்.
- **மொழி குறிப்பிட்டது**. சில பயன்பாடுகள் பல மொழிகளை ஆதரிக்கின்றன, ஆனால் இயல்பாக பயன்பாடு ஒரு குறிப்பிட்ட மொழிக்காக உருவாக்கப்பட்டுள்ளது, மேலும் நீங்கள் கூடுதல் மொழி ஆதரவைச் சேர்க்க முடியும்.

### உரை உருவாக்க பயன்பாடுகளின் நன்மைகள்

அப்படியென்றால், ஒரு உரை உருவாக்க பயன்பாடு எப்படி வேறுபடுகிறது?

ஒரு உரை உருவாக்க பயன்பாட்டில், நீங்கள் அதிக சுதந்திரம் பெறுகிறீர்கள்; நீங்கள் ஒரு கமாண்ட்களின் தொகுப்புக்கு அல்லது ஒரு குறிப்பிட்ட உள்ளீடு மொழிக்கு மட்டுமே வரையறுக்கப்படவில்லை. அதற்கு பதிலாக, நீங்கள் இயற்கை மொழியைப் பயன்படுத்தி பயன்பாட்டுடன் தொடர்பு கொள்ளலாம். மேலும், நீங்கள் ஏற்கனவே ஒரு பெரிய தகவல் தொகுப்பில் பயிற்சி பெற்ற தரவுத் மூலத்துடன் தொடர்பு கொள்ளுகிறீர்கள், ஆனால் ஒரு பாரம்பரிய பயன்பாடு ஒரு தரவுத்தொகுப்பில் உள்ளவற்றில் மட்டுமே வரையறுக்கப்பட்டிருக்கலாம்.

### உரை உருவாக்க பயன்பாட்டைப் பயன்படுத்தி என்ன உருவாக்க முடியும்?

நீங்கள் பலவற்றை உருவாக்க முடியும். உதாரணமாக:

- **ஒரு chatbot**. உங்கள் நிறுவனம் மற்றும் அதன் தயாரிப்புகள் போன்ற தலைப்புகள் குறித்து கேள்விகளுக்கு பதிலளிக்கும் ஒரு chatbot ஒரு நல்ல பொருத்தமாக இருக்கலாம்.
- **உதவியாளர்**. LLMகள் உரையை சுருக்குதல், உரையிலிருந்து தகவல்களைப் பெறுதல், ரெஸ்யூமே போன்ற உரைகளை உருவாக்குதல் போன்றவற்றில் சிறந்தவை.
- **கோடு உதவியாளர்**. நீங்கள் பயன்படுத்தும் மொழி மாதிரியைப் பொறுத்து, நீங்கள் கோடு எழுத உதவும் ஒரு கோடு உதவியாளரை உருவாக்க முடியும். உதாரணமாக, GitHub Copilot மற்றும் ChatGPT போன்ற தயாரிப்புகளைப் பயன்படுத்தி கோடு எழுத உதவ முடியும்.

## நான் எப்படி தொடங்கலாம்?

நீங்கள் ஒரு LLM உடன் ஒருங்கிணைக்க ஒரு வழியை கண்டுபிடிக்க வேண்டும், இது பொதுவாக இரண்டு அணுகுமுறைகளை உள்ளடக்கியது:

- **API பயன்படுத்துதல்**. இங்கு நீங்கள் உங்கள் ப்ராம்ப்டுடன் வலை கோரிக்கைகளை உருவாக்கி, உருவாக்கப்பட்ட உரையை திரும்பப் பெறுகிறீர்கள்.
- **நூலகம் பயன்படுத்துதல்**. நூலகங்கள் API அழைப்புகளை மூடிவைத்து அவற்றைப் பயன்படுத்த எளிதாக்குகின்றன.

## நூலகங்கள்/SDKகள்

LLMகளுடன் வேலை செய்ய சில பிரபலமான நூலகங்கள் உள்ளன:

- **openai**, இந்த நூலகம் உங்கள் மாதிரியை இணைத்து ப்ராம்ப்ட்களை அனுப்ப எளிதாக்குகிறது.

பிறகு, மேல்நிலையான செயல்பாடுகளை வழங்கும் நூலகங்கள் உள்ளன:

- **Langchain**. Langchain மிகவும் பிரபலமானது மற்றும் Python-ஐ ஆதரிக்கிறது.
- **Semantic Kernel**. Semantic Kernel என்பது Microsoft உருவாக்கிய ஒரு நூலகம், இது C#, Python, மற்றும் Java மொழிகளை ஆதரிக்கிறது.

## openai பயன்படுத்தி முதல் பயன்பாடு

நாம் எவ்வாறு முதல் பயன்பாட்டை உருவாக்கலாம், எந்த நூலகங்கள் தேவை, எவ்வளவு தேவை போன்றவற்றைப் பார்ப்போம்.

### openai நிறுவுதல்

  > [!NOTE] இந்த படி Codespaces அல்லது Devcontainer-ல் இந்த நோட்புக் இயக்கும்போது தேவையில்லை

OpenAI அல்லது Azure OpenAI உடன் தொடர்பு கொள்ள பல நூலகங்கள் உள்ளன. C#, Python, JavaScript, Java மற்றும் பல போன்ற பல நிரலாக்க மொழிகளைப் பயன்படுத்த முடியும்.  
நாம் `openai` Python நூலகத்தைத் தேர்ந்தெடுத்துள்ளோம், எனவே `pip` பயன்படுத்தி அதை நிறுவுவோம்.

```bash
pip install openai
```

இந்த நோட்புக் Codespaces அல்லது Dev Container-ல் இயக்கப்படவில்லை என்றால், உங்கள் கணினியில் [Python](https://www.python.org/) ஐ நிறுவவும்.

### ஒரு resource உருவாக்குதல்

நீங்கள் இதுவரை செய்யவில்லை என்றால், பின்வரும் படிகளைச் செய்ய வேண்டும்:

- Azure-ல் ஒரு கணக்கை உருவாக்கவும் <https://azure.microsoft.com/free/>.
- Azure OpenAI-க்கு அணுகலைப் பெறவும். <https://learn.microsoft.com/azure/ai-services/openai/overview#how-do-i-get-access-to-azure-openai> சென்று அணுகலைக் கோரவும்.

  > [!NOTE]
  > இந்த ஆவணத்தை எழுதும் நேரத்தில், Azure OpenAI-க்கு அணுகலைப் பெற நீங்கள் விண்ணப்பிக்க வேண்டும்.

- Azure OpenAI Service resource ஒன்றை உருவாக்கவும். [resource உருவாக்க](https://learn.microsoft.com/azure/ai-services/openai/how-to/create-resource?pivots=web-portal&WT.mc_id=academic-105485-koreyst) எப்படி என்பதைப் பற்றிய இந்த வழிகாட்டியைப் பார்க்கவும்.

### API key மற்றும் endpoint கண்டறிதல்

இந்த நேரத்தில், உங்கள் `openai` நூலகத்திற்கு எந்த API key ஐப் பயன்படுத்த வேண்டும் என்பதைச் சொல்ல வேண்டும். உங்கள் API key ஐ கண்டறிய, Azure OpenAI resource இன் "Keys and Endpoint" பிரிவுக்கு சென்று "Key 1" மதிப்பை நகலெடுக்கவும்.

  ![Azure Portal இல் Keys and Endpoint resource blade](https://learn.microsoft.com/azure/ai-services/openai/media/quickstarts/endpoint.png?WT.mc_id=academic-105485-koreyst)

இப்போது நீங்கள் இந்த தகவலை நகலெடுத்துள்ளீர்கள், நூலகங்களுக்கு அதை பயன்படுத்த சொல்லுவோம்.

> [!NOTE]
> உங்கள் API key ஐ உங்கள் கோடில் இருந்து பிரிக்க வேண்டும். நீங்கள் சூழல் மாறிகள் (environment variables) பயன்படுத்தி அதைச் செய்யலாம்.
> - உங்கள் .env கோப்பில் `AZURE_OPENAI_API_KEY` சூழல் மாறியை உங்கள் API key ஆக அமைக்கவும். இந்த பாடத்திட்டத்தின் முந்தைய பயிற்சிகளை நீங்கள் முடித்திருந்தால், நீங்கள் அனைத்தும் அமைத்துவிட்டீர்கள்.

### Azure அமைப்பை அமைத்தல்

நீங்கள் Azure OpenAI ஐ பயன்படுத்தினால், அமைப்பை அமைப்பது எப்படி என்பதை இங்கே பாருங்கள்:

```python
client = AzureOpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],  
  api_version = "2023-10-01-preview"
  azure_endpoint = os.environ('AZURE_OPENAI_ENDPOINT')
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']
```

மேலே நாம் பின்வருவனவற்றை அமைக்கிறோம்:

- `api_key`, இது Azure Portal-ல் காணப்படும் உங்கள் API key.
- `api_version`, இது நீங்கள் பயன்படுத்த விரும்பும் API இன் பதிப்பு. இந்த ஆவணத்தை எழுதும் நேரத்தில், சமீபத்திய பதிப்பு `2023-10-01-preview`.
- `azure_endpoint`, இது API இன் endpoint. உங்கள் API key க்கு அருகில் Azure Portal-ல் இதை நீங்கள் காணலாம்.

## பயிற்சி - ஒரு சமையல் குறிப்புகள் உருவாக்குபவை உருவாக்குங்கள்

சிறப்பு, இப்போது நாம் விஷயங்களை எவ்வாறு மேம்படுத்தலாம் என்று பார்ப்போம். விஷயங்களை மேம்படுத்த, குறியீடு நெகிழ்வானதாக இருக்க வேண்டும், எனவே பொருட்கள் மற்றும் சமையல் எண்ணிக்கையை மேம்படுத்தவும் மாற்றவும் முடியும்.


1. நாம் கீழே உள்ள முறையில் குறியீட்டை மாற்றுவோம்:


````python
import os
from openai import AzureOpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = AzureOpenAI(
  azure_endpoint = os.environ["AZURE_OPENAI_ENDPOINT"], 
  api_key=os.environ['AZURE_OPENAI_API_KEY'],  
  api_version = "2023-10-01-preview"
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"
messages = [{"role": "user", "content": prompt}]  

# make completion
completion = client.chat.completions.create(model=deployment, messages=messages, max_tokens=600)

# print response
print(completion.choices[0].message.content)
````
