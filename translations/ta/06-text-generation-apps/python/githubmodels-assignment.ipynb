# உரை உருவாக்க பயன்பாடுகளை உருவாக்குதல்

இந்த பாடத்திட்டத்தின் மூலம் நீங்கள் இதுவரை முக்கியமான கருத்துக்களைப் பார்த்துள்ளீர்கள், உதாரணமாக "ப்ராம்ப்ட்" மற்றும் "ப்ராம்ப்ட் இன்ஜினியரிங்" எனப்படும் ஒரு முழு துறை. ChatGPT, Office 365, Microsoft Power Platform போன்ற பல கருவிகளைப் பயன்படுத்தி, ப்ராம்ப்ட்களை பயன்படுத்தி ஒரு செயல்பாட்டை நிறைவேற்ற முடியும்.

இந்த அனுபவத்தை ஒரு பயன்பாட்டில் சேர்க்க நீங்கள் ப்ராம்ப்ட்கள், நிறைவுகள் போன்ற கருத்துகளைப் புரிந்துகொண்டு, வேலை செய்ய ஒரு நூலகத்தைத் தேர்ந்தெடுக்க வேண்டும். இதை நீங்கள் இந்த அத்தியாயத்தில் கற்றுக்கொள்வீர்கள்.

## அறிமுகம்

இந்த அத்தியாயத்தில், நீங்கள்:

- openai நூலகம் மற்றும் அதன் முக்கியமான கருத்துகளைப் பற்றி கற்றுக்கொள்வீர்கள்.
- openai பயன்படுத்தி ஒரு உரை உருவாக்க பயன்பாட்டை உருவாக்குவீர்கள்.
- ப்ராம்ப்ட், வெப்பநிலை, மற்றும் டோக்கன்கள் போன்ற கருத்துகளைப் பயன்படுத்தி ஒரு உரை உருவாக்க பயன்பாட்டை உருவாக்குவது எப்படி என்பதைப் புரிந்துகொள்வீர்கள்.

## கற்றல் இலக்குகள்

இந்த பாடத்தின் முடிவில், நீங்கள்:

- ஒரு உரை உருவாக்க பயன்பாடு என்ன என்பதை விளக்க முடியும்.
- openai பயன்படுத்தி ஒரு உரை உருவாக்க பயன்பாட்டை உருவாக்க முடியும்.
- உங்கள் பயன்பாட்டை அதிகமாக அல்லது குறைவாக டோக்கன்களைப் பயன்படுத்தவும், வெப்பநிலையை மாற்றவும் அமைக்க முடியும், பல்வேறு வெளியீடுகளுக்காக.

## உரை உருவாக்க பயன்பாடு என்ன?

சாதாரணமாக, நீங்கள் ஒரு பயன்பாட்டை உருவாக்கும்போது, அது பின்வருவன போன்ற ஒரு இடைமுகத்தை கொண்டிருக்கும்:

- **கமாண்டு அடிப்படையிலானது**. Console பயன்பாடுகள் பொதுவாக நீங்கள் ஒரு கமாண்டை டைப் செய்து ஒரு பணியைச் செய்யும் பயன்பாடுகள். உதாரணமாக, `git` என்பது ஒரு கமாண்டு அடிப்படையிலான பயன்பாடு.
- **பயனர் இடைமுகம் (UI)**. சில பயன்பாடுகள் கிராபிகல் பயனர் இடைமுகங்களை (GUIs) கொண்டிருக்கும், இதில் நீங்கள் பட்டன்களை கிளிக் செய்யலாம், உரை உள்ளிடலாம், விருப்பங்களைத் தேர்ந்தெடுக்கலாம் மற்றும் பல.

### Console மற்றும் UI பயன்பாடுகள் வரையறுக்கப்பட்டவை

ஒரு கமாண்டு அடிப்படையிலான பயன்பாட்டுடன் ஒப்பிடுக:

- **இது வரையறுக்கப்பட்டது**. நீங்கள் எந்த கமாண்டையும் டைப் செய்ய முடியாது, பயன்பாடு ஆதரிக்கும் கமாண்டுகளையே மட்டும் டைப் செய்ய முடியும்.
- **மொழி குறிப்பிட்டது**. சில பயன்பாடுகள் பல மொழிகளை ஆதரிக்கின்றன, ஆனால் இயல்பாக பயன்பாடு ஒரு குறிப்பிட்ட மொழிக்காக உருவாக்கப்பட்டுள்ளது, கூடுதல் மொழி ஆதரவைச் சேர்க்க முடிந்தாலும்.

### உரை உருவாக்க பயன்பாடுகளின் நன்மைகள்

அப்படியென்றால், ஒரு உரை உருவாக்க பயன்பாடு எப்படி வேறுபடுகிறது?

ஒரு உரை உருவாக்க பயன்பாட்டில், நீங்கள் அதிக சுதந்திரம் பெறுகிறீர்கள், ஒரு கமாண்டுகளின் தொகுப்புக்கு அல்லது ஒரு குறிப்பிட்ட உள்ளீடு மொழிக்கு மட்டுமே வரையறுக்கப்படவில்லை. அதற்கு பதிலாக, நீங்கள் இயற்கை மொழியைப் பயன்படுத்தி பயன்பாட்டுடன் தொடர்பு கொள்ளலாம். மேலும், நீங்கள் ஏற்கனவே ஒரு பெரிய தகவல் தொகுப்பில் பயிற்சி பெற்ற தரவின் மூலம் தொடர்பு கொள்கிறீர்கள், ஆனால் ஒரு பாரம்பரிய பயன்பாடு ஒரு தரவுத்தொகுப்பில் உள்ளவற்றில் மட்டுமே வரையறுக்கப்பட்டிருக்கலாம்.

### உரை உருவாக்க பயன்பாட்டைப் பயன்படுத்தி என்ன உருவாக்க முடியும்?

நீங்கள் பலவற்றை உருவாக்க முடியும். உதாரணமாக:

- **ஒரு chatbot**. உங்கள் நிறுவனம் மற்றும் அதன் தயாரிப்புகள் போன்ற தலைப்புகள் குறித்து கேள்விகளுக்கு பதிலளிக்கும் ஒரு chatbot ஒரு நல்ல பொருத்தமாக இருக்கலாம்.
- **உதவியாளர்**. LLMs உரையை சுருக்குதல், உரையிலிருந்து தகவல்களைப் பெறுதல், ரெஸ்யூமே போன்ற உரைகளை உருவாக்குதல் போன்றவற்றில் சிறந்தவை.
- **கோடு உதவியாளர்**. நீங்கள் பயன்படுத்தும் மொழி மாதிரியைப் பொறுத்து, கோடு எழுத உதவும் ஒரு கோடு உதவியாளரை உருவாக்க முடியும். உதாரணமாக, GitHub Copilot மற்றும் ChatGPT போன்ற தயாரிப்புகளைப் பயன்படுத்தி கோடு எழுத உதவ முடியும்.

## நான் எப்படி தொடங்கலாம்?

நீங்கள் ஒரு LLM உடன் ஒருங்கிணைக்க ஒரு வழியை கண்டுபிடிக்க வேண்டும், இது பொதுவாக இரண்டு அணுகுமுறைகளை உள்ளடக்கியது:

- **API ஐப் பயன்படுத்துதல்**. இங்கு நீங்கள் உங்கள் ப்ராம்ப்டுடன் வலை கோரிக்கைகளை உருவாக்கி, உருவாக்கப்பட்ட உரையை திரும்பப் பெறுகிறீர்கள்.
- **நூலகத்தைப் பயன்படுத்துதல்**. நூலகங்கள் API அழைப்புகளை மூடிவைத்து, அவற்றைப் பயன்படுத்த எளிதாக்குகின்றன.

## பயிற்சி - ஒரு சமையல் குறிப்புகள் உருவாக்கியை உருவாக்குங்கள்

இப்போது நாங்கள் ஒரு சூழலை விளையாடியுள்ளோம், காட்டப்பட்ட சூழலுக்கு பொருந்தும் குறியீட்டை எழுதுவோம். இதைச் செய்ய, கீழே உள்ள படிகளைப் பின்பற்றுங்கள்:

1. உள்ளமைந்த கோப்பை தொடக்கமாக பயன்படுத்துங்கள்
1. ஒரு `prompt` மாறியை உருவாக்கி, கீழே உள்ள மாதிரிக் குறியீட்டைப் மாற்றுங்கள்:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

token = os.environ["GITHUB_TOKEN"]
endpoint = "https://models.inference.ai.azure.com"

model_name = "gpt-4o"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

1. கீழே உள்ள முறையில் குறியீட்டை மாற்றுவோம்:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

token = os.environ["GITHUB_TOKEN"]
endpoint = "https://models.inference.ai.azure.com"

model_name = "gpt-4o"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````
