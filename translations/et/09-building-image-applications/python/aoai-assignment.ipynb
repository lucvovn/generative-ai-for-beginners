## Mis on DALL-E ja Midjourney?

[DALL-E](https://openai.com/dall-e-2?WT.mc_id=academic-105485-koreyst) ja [Midjourney](https://www.midjourney.com/?WT.mc_id=academic-105485-koreyst) on kaks kõige populaarsemat pildigeneratsiooni mudelit, mis võimaldavad luua pilte promptide abil.

### DALL-E

Alustame DALL-E-st, mis on generatiivse tehisintellekti mudel, mis loob pilte tekstikirjeldustest.

> [DALL-E on kahe mudeli, CLIP ja hajutatud tähelepanu, kombinatsioon](https://towardsdatascience.com/openais-dall-e-and-clip-101-a-brief-introduction-3a4367280d4e?WT.mc_id=academic-105485-koreyst).

- **CLIP** on mudel, mis loob sisenditest (piltidest ja tekstist) numbrilisi esindusi ehk embeddings'eid.

- **Hajutatud tähelepanu** on mudel, mis loob embeddings'itest pilte. DALL-E on treenitud piltide ja tekstide andmestikul ning seda saab kasutada piltide loomiseks tekstikirjeldustest. Näiteks saab DALL-E abil luua pilte, kus kass kannab mütsi või koeral on mohawk.

### Midjourney

Midjourney töötab sarnaselt DALL-E-le, luues pilte tekstipromptide abil. Midjourney abil saab samuti luua pilte, kasutades prompt'e nagu "kass mütsiga" või "koer mohawk'iga".

## Kuidas DALL-E ja Midjourney töötavad

Esiteks, [DALL-E](https://arxiv.org/pdf/2102.12092.pdf?WT.mc_id=academic-105485-koreyst). DALL-E on generatiivse tehisintellekti mudel, mis põhineb transformer-arhitektuuril ja kasutab *autoregressiivset transformerit*.

*Autoregressiivne transformer* määratleb, kuidas mudel loob pilte tekstikirjeldustest – see genereerib ühe piksli korraga ja kasutab loodud piksleid järgmise piksli loomiseks. Protsess läbib mitmeid kihte närvivõrgus, kuni pilt on valmis.

Selle protsessi abil suudab DALL-E kontrollida pildi omadusi, objekte, karakteristikuid ja palju muud. Kuid DALL-E 2 ja 3 pakuvad veelgi suuremat kontrolli loodud pildi üle.


## Esimese pildigeneratsiooni rakenduse loomine

Mis on vajalik pildigeneratsiooni rakenduse loomiseks? Vajate järgmisi teeke:

- **python-dotenv**, tungivalt soovitatav kasutada seda teeki, et hoida oma konfidentsiaalsed andmed *.env* failis eemal koodist.
- **openai**, see teek võimaldab teil suhelda OpenAI API-ga.
- **pillow**, piltidega töötamiseks Pythonis.
- **requests**, HTTP-päringute tegemiseks.

1. Looge fail *.env* järgmise sisuga:

    ```text
    AZURE_OPENAI_ENDPOINT=<your endpoint>
    AZURE_OPENAI_API_KEY=<your key>
    ```

    Selle teabe leiate Azure Portaalist oma ressursi "Keys and Endpoint" sektsioonist.


1. Kogu ülaltoodud teegid faili nimega *requirements.txt*, näiteks nii:

    ```text
    python-dotenv
    openai
    pillow
    requests
    ```


1. Järgmiseks loo virtuaalne keskkond ja paigalda teegid:


`````python
# create virtual env
! python3 -m venv venv
# activate environment
! source venv/bin/activate
# install libraries
# pip install -r requirements.txt, if using a requirements.txt file 
! pip install python-dotenv openai pillow requests
`````

> [!NOTE]
> Windowsi jaoks kasutage virtuaalse keskkonna loomiseks ja aktiveerimiseks järgmisi käske:

    ```bash

    venv\Scripts\activate.bat
    ````

1. Lisage järgmine kood faili nimega *app.py*:

    ```python
    import openai
    import os
    import requests
    from PIL import Image
    import dotenv
    
    # import dotenv
    dotenv.load_dotenv()
    
    # Get endpoint and key from environment variables
    openai.api_base = os.environ['AZURE_OPENAI_ENDPOINT']
    openai.api_key = os.environ['AZURE_OPENAI_API_KEY']     
    
    # Assign the API version (DALL-E is currently supported for the 2023-06-01-preview API version only)
    openai.api_version = '2023-06-01-preview'
    openai.api_type = 'azure'
    
    
    try:
        # Create an image by using the image generation API
        generation_response = openai.Image.create(
            prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # Enter your prompt text here
            size='1024x1024',
            n=2,
            temperature=0,
        )
        # Set the directory for the stored image
        image_dir = os.path.join(os.curdir, 'images')
    
        # If the directory doesn't exist, create it
        if not os.path.isdir(image_dir):
            os.mkdir(image_dir)
    
        # Initialize the image path (note the filetype should be png)
        image_path = os.path.join(image_dir, 'generated-image.png')
    
        # Retrieve the generated image
        image_url = generation_response["data"][0]["url"]  # extract image URL from response
        generated_image = requests.get(image_url).content  # download the image
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)
    
        # Display the image in the default image viewer
        image = Image.open(image_path)
        image.show()
    
    # catch exceptions
    except openai.InvalidRequestError as err:
        print(err)

    ```

### Rohkem detaile pildi genereerimise kohta

Vaatame pildi genereerimise koodi lähemalt:

```python
generation_response = openai.Image.create(
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # Enter your prompt text here
        size='1024x1024',
        n=2,
        temperature=0,
    )
```

- **prompt** on tekstiline sisend, mida kasutatakse pildi genereerimiseks. Antud juhul kasutame sisendit "Jänes hobuse seljas, hoides pulgakommi, udusel aasale, kus kasvavad nartsissid".
- **size** on genereeritud pildi suurus. Antud juhul genereerime pildi suurusega 1024x1024 pikslit.
- **n** on genereeritavate piltide arv. Antud juhul genereerime kaks pilti.
- **temperature** on parameeter, mis kontrollib generatiivse tehisintellekti mudeli väljundi juhuslikkust. Temperatuur on väärtus vahemikus 0 kuni 1, kus 0 tähendab, et väljund on deterministlik, ja 1 tähendab, et väljund on juhuslik. Vaikeväärtus on 0.7.

Piltidega saab teha veel palju muud, mida käsitleme järgmises jaotises.

## Piltide genereerimise täiendavad võimalused

Olete seni näinud, kuidas saime Pythonis mõne koodireaga pildi genereerida. Siiski on piltidega võimalik teha veel palju muud.

Saate teha ka järgmist:

- **Teha muudatusi**. Andes olemasoleva pildi, maski ja sisendi, saate pilti muuta. Näiteks saate lisada midagi pildi teatud osale. Kujutage ette meie jänese pilti – saate lisada jänesele mütsi. Seda saab teha, andes pildi, maski (mis määrab ala, mida muuta) ja tekstilise sisendi, mis ütleb, mida teha.

    ```python
    response = openai.Image.create_edit(
      image=open("base_image.png", "rb"),
      mask=open("mask.png", "rb"),
      prompt="An image of a rabbit with a hat on its head.",
      n=1,
      size="1024x1024"
    )
    image_url = response['data'][0]['url']
    ```

    Algne pilt sisaldaks ainult jänest, kuid lõplikul pildil oleks jänesel müts.
